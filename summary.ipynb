```python
import pandas as pd

# Helper function to clean and convert the 'Accuracy' column
def clean_accuracy_column(df):
    df['Accuracy'] = df['Accuracy'].str.rstrip('%').astype(float) / 100  # Remove '%' and convert to float

# Manually load the CSV files
BA = pd.read_csv("result/bright_setting/brightAug_log.csv")
BO = pd.read_csv("result/bright_setting/brightOg_log.csv")
DA = pd.read_csv("result/dark_setting/darkAug_log.csv")
DO = pd.read_csv("result/dark_setting/darkOg_log.csv")
NA = pd.read_csv("result/normal_setting/normalAug_log.csv")
NO = pd.read_csv("result/normal_setting/normalOg_log.csv")

# List of CSVs and their labels for easier summarization
csv_files = [(BA, "brightAug_log.csv"),
             (BO, "brightOg_log.csv"),
             (DA, "darkAug_log.csv"),
             (DO, "darkOg_log.csv"),
             (NA, "normalAug_log.csv"),
             (NO, "normalOg_log.csv")]

# Summarize accuracy column for each CSV
for df, label in csv_files:
    if 'Accuracy' in df.columns:
        clean_accuracy_column(df)  # Clean and convert accuracy values to numeric
        print(f"Summary for {label}:")

        # Calculating max, min, mean, std, mode for the 'Accuracy' column
        max_value = df['Accuracy'].max()
        min_value = df['Accuracy'].min()
        mean_value = df['Accuracy'].mean()
        std_value = df['Accuracy'].std()
        mode_value = df['Accuracy'].mode()[0] if not df['Accuracy'].mode().empty else 'No mode'

        print(f"Max: {max_value}")
        print(f"Min: {min_value}")
        print(f"Mean: {mean_value}")
        print(f"Std: {std_value}")
        print(f"Mode: {mode_value}")
        print("\n" + "="*50 + "\n")
    else:
        print(f"Error: 'Accuracy' column not found in {label}\n")

```

```python
import pandas as pd

# Helper function to clean and convert the 'Accuracy' column
def clean_accuracy_column(df):
    df['Accuracy'] = df['Accuracy'].str.rstrip('%').astype(float) / 100  # Remove '%' and convert to float

# Manually load the CSV files
BA = pd.read_csv("result/bright_setting/brightAug_log.csv")
BO = pd.read_csv("result/bright_setting/brightOg_log.csv")
DA = pd.read_csv("result/dark_setting/darkAug_log.csv")
DO = pd.read_csv("result/dark_setting/darkOg_log.csv")
NA = pd.read_csv("result/normal_setting/normalAug_log.csv")
NO = pd.read_csv("result/normal_setting/normalOg_log.csv")

# List of CSVs and their labels for easier summarization
csv_files = [(BA, "brightAug_log.csv"),
             (BO, "brightOg_log.csv"),
             (DA, "darkAug_log.csv"),
             (DO, "darkOg_log.csv"),
             (NA, "normalAug_log.csv"),
             (NO, "normalOg_log.csv")]

# Summarize accuracy column for each CSV
for df, label in csv_files:
    if 'Accuracy' in df.columns:
        clean_accuracy_column(df)  # Clean and convert accuracy values to numeric
        print(f"Summary for {label}:")

        # Length of the 'Accuracy' column
        row_count = len(df)

        # Calculating max, min, mean, std, mode for the 'Accuracy' column
        max_value = df['Accuracy'].max()
        min_value = df['Accuracy'].min()
        mean_value = df['Accuracy'].mean()
        std_value = df['Accuracy'].std()
        mode_value = df['Accuracy'].mode()[0] if not df['Accuracy'].mode().empty else 'No mode'

        print(f"Length: {row_count}")
        print(f"Max: {max_value}")
        print(f"Min: {min_value}")
        print(f"Mean: {mean_value}")
        print(f"Std: {std_value}")
        print(f"Mode: {mode_value}")
        print("\n" + "="*50 + "\n")
    else:
        print(f"Error: 'Accuracy' column not found in {label}\n")
```
